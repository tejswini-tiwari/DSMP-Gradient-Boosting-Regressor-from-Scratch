```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import train_test_split
```

```python
titanic  = pd.read_csv("train.csv")
```

```python
# titanic.isna().sum()
```

```python
titanic.drop(["Cabin"], axis=1, inplace=True)
```

```python
titanic.dropna(inplace=True)
```

```python
cols = ["Pclass", "Sex","Age", "Fare"]
df=titanic[cols]
```

```python
labelEncoder = LabelEncoder()
df['Sex'] = labelEncoder.fit_transform(df["Sex"])
```

```python
data=df.sample(140, random_state=42)
```

```python
data.sort_index(inplace=True)
```

```python
df.drop(data.index, inplace=True)
```

```python
N=df.shape[0]
Models = 10
def Loss(yi, yipred):
    return np.square(yi-yipred)
```

```python
N
```

```python
df["f0x"]=df["Fare"].mean()
models=["f0x"]
```

```python
def train_model(m):
    pseudo_res=f"ri{m}"
    prev_model_out = f"f{m-1}x"
    df[pseudo_res]=df["Fare"]-df[prev_model_out]

    model = f"dtree{m}"
    model = DecisionTreeRegressor(max_depth=3)
    model.fit(df.iloc[:,1:4], df[pseudo_res])
    models.append(model)

    new_model_out = f"f{m}x"
    df[new_model_out]= df[prev_model_out]+model.predict(df.iloc[:,1:4])
```

```python
# train_model(1)
```

```python
def fit():
    for m in range(1,Models):
        train_model(m)
```

```python
fit()
```

```python
def predict(data):
    data["f0x"]=df["Fare"].mean()
    for m in range(1,Models):
        prev_model_out = f"f{m-1}x"
          
        model = models[m]

        new_model_out = f"f{m}x"
        data[new_model_out]= data[prev_model_out]+model.predict(data.iloc[:,1:4])
```
